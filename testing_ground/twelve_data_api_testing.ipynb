```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

# You will need an API key and an env file for that:
TWELVE_DATA_API_KEY = os.getenv('TwelveDataAPI')
```

```python
get_daily_data = f'https://api.twelvedata.com/time_series?start_date=2020-05-06&outputsize=5&symbol=aapl&interval=1day&apikey={TWELVE_DATA_API_KEY}'
response = requests.get(get_daily_data)
```

```python
five_days = response.json()
five_days
```

```python
five_days['values']
```

```python
get_hourly_data = f'https://api.twelvedata.com/time_series?start_date=2020-05-06&outputsize=5&symbol=aapl&interval=1h&apikey={TWELVE_DATA_API_KEY}'
response = requests.get(get_hourly_data)
```
