```python
import requests
import pandas as pd
import altair as alt
import json
```

```python
# Task: Get Conversion Rates for One Week EUR > USD
# * Daterange: https://api.frankfurter.dev/v1/2000-01-01..2000-12-31
# * Daterange to present: https://api.frankfurter.dev/v1/2025-05-15..
# * Specific base and target: https://api.frankfurter.dev/v1/1999-01-04?base=USD&symbols=EUR
```

```python
some_time_eur_to_dollar = 'https://api.frankfurter.dev/v1/2025-04-15..?base=EUR&symbols=USD'
response = requests.get(some_time_eur_to_dollar)
```

```python
response.json()
```

```python
five_years_eur_to_usd = 'https://api.frankfurter.dev/v1/2020-05-22..?base=EUR&symbols=USD'
response = requests.get(five_years_eur_to_usd)
```

```python
five_years_data = response.json()
five_years_data
```

```python
five_years_frame = pd.DataFrame.from_dict(five_years_data['rates'], orient='index')
five_years_frame
```

```python
len(five_years_frame)
```

```python
len(five_years_frame) / (365 -55*2)
```

```python
five_years_frame.plot(
    title='Five Years EUR > USD Exchange Rate ',
    legend=False,
    rot=45
);
```
