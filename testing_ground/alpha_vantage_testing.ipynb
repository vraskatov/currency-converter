```python
import requests
import pandas as pd
import altair as alt
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

# You will need an API key and an env file for that:
ALPHA_VANTAGE_KEY = os.getenv('AlphaVantageAPI')
```

```python
test_call = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&apikey={ALPHA_VANTAGE_KEY}'
response = requests.get(test_call)
```

```python
response.json()
```

```python
usd_call = f'https://www.alphavantage.co/query?function=CURRENCY_EXCHANGE_RATE&from_currency=EUR&to_currency=USD&apikey={ALPHA_VANTAGE_KEY}'
response = requests.get(usd_call)
```

```python
response.json()
```

```python
eur_to_usd_full = f'https://www.alphavantage.co/query?function=FX_DAILY&from_symbol=EUR&to_symbol=USD&apikey={ALPHA_VANTAGE_KEY}'
response = requests.get(eur_to_usd_full)
```

```python
result = response.json()
result
```

```python
result['Time Series FX (Daily)']
```

```python
eur_to_usd_full_csv = f'https://www.alphavantage.co/query?function=FX_DAILY&from_symbol=EUR&to_symbol=USD&apikey={ALPHA_VANTAGE_KEY}&datatype=csv'
response = requests.get(eur_to_usd_full_csv)
```

```python
response.text
```

```python
with open('daily_data.csv', "w") as f:
    f.writelines(response.text)
```

```python
daily_100 = pd.read_csv('daily_data.csv')
daily_100.head()
```

```python
daily_100.shape
```

```python
ax = daily_100.plot(x='timestamp', y='open')
ax.set_title('100 days Exchange Rate EUR > USD');
```

```python
# Altair is part of Streamlit, so let's try that as well:
y_min = daily_100['open'].min()
y_max = daily_100['open'].max()
buffer = (y_max - y_min) * 0.05

lower_bound = y_min - buffer
print(lower_bound)
upper_bound = y_max + buffer
print(upper_bound)
alt.Chart(daily_100).mark_line().encode(
    x='timestamp',
    y=alt.Y(
        'open',
        scale=alt.Scale(domain=[lower_bound, upper_bound])
    )
)
```
